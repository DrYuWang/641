```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
X = pd.read_csv("dataset_1.csv").values
y = X[:, 2]
X = X[:, :2]
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2)
R = GaussianNB()
R.fit(X_tr, y_tr)
y_p = R.predict(X_te)
y_p.shape
X_te.shape
```

```python
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(X_te[y_te == 0, 0], X_te[y_te == 0, 1], 'bo')
plt.plot(X_te[y_te == 1, 0], X_te[y_te == 1, 1], 'ro')
plt.show()
```

```python
u = np.arange(-10, 10, 0.1)
v = np.arange(-10, 10, 0.1)
U, V = np.meshgrid(u, v)
Z = []
for i in range(200):
    W = np.array([U[:, i], V[:, i]]).T
#     print(W.shape)
    z = R.predict(W)
    Z.append(z)
ZZ = np.array(Z).T
print(ZZ.shape)

plt.figure(figsize=(6,6))
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.plot(X_te[y_te == 0, 0], X_te[y_te == 0, 1], 'bo')
plt.plot(X_te[y_te == 1, 0], X_te[y_te == 1, 1], 'ro')
plt.contourf(u, v, ZZ)
plt.show()
```
