```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def likelihood(data):
    x = data[:,0]
    y = data[:,1]
    def f(t):
        factors = np.power(sigmoid(t*x), y)*np.power(1-sigmoid(t*x), 1-y)
        return np.prod(factors)
    return f
```

```python
# data = np.array([[-1, 1], [0, 0], [0.5, 1], [1, 0]])
# failures = np.array([np.random.uniform(-0.1, 0, size=5), np.zeros((5,))]).T
# successes = np.array([np.random.uniform(0, 0.1, size=5), np.ones((5,))]).T
# data = np.vstack((failures, successes))
# data = np.vstack((data, np.array([[0.1, 0]])))
data = np.array([np.random.uniform(-1, 1, size=10), np.random.randint(2, size=10)])
f = likelihood(data)

T = np.linspace(-100, 20, 100)
# plt.plot(data[:,0], data[:,1], 'bo')
plt.plot(T, [f(t) for t in T], 'r')
plt.show()
```

```python
a = np.linspace(-15, 15, 100)
b = np.linspace(-5, 5, 100)
A, B = np.meshgrid(a, b)
C = -np.log((1 - sigmoid(A + B))*sigmoid(A + 2*B)*(1 - sigmoid(A + 3*B)))
```

```python
plt.contour(a, b, C, np.log(np.arange(0.01, 0.14, 0.01)))
```
