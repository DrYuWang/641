```python
from sklearn.linear_model import RidgeCV
from sklearn.datasets import load_diabetes
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 70
p = 20
x = np.random.normal(size=(n, p))
y = np.random.normal(size=n)
alphas = np.arange(0.01, 1000, 1)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(x, y)
```

```python
cvs = np.mean(clf.cv_values_, axis=0)
plt.plot(alphas, cvs)
plt.show()
print(clf.alpha_)
```

```python
_, s, __ = np.linalg.svd(x.transpose() @ x)
```

```python
plt.plot(range(len(s)), s)
plt.show()
```

```python
X, y = load_diabetes(return_X_y=True)
alphas = np.arange(0.01, 2, 0.01)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(X, y)
cvs = np.mean(clf.cv_values_, axis=0)
plt.plot(alphas, cvs)
plt.show()
```
