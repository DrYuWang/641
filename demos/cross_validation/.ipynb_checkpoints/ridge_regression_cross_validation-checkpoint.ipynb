```python
from sklearn import linear_model
from sklearn.datasets import load_diabetes, load_boston
from sklearn.linear_model import RidgeCV
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 10
p = 10

x = np.random.normal(0, 1, size=(n, p))
y = x[:,0] + np.random.normal(0, 1, size=10)
```

```python
xt = x.transpose()
b = np.linalg.inv(xt @ x) @ xt @ y
print(b)
```

```python
fit = linear_model.LinearRegression(fit_intercept=False).fit(x, y)
print(fit.coef_)
```

```python
alphas = np.arange(0,0.1,0.0001)
coeffs = []
for alpha in alphas:
    fit = linear_model.Ridge(fit_intercept=False, alpha=alpha).fit(x, y)
    coeffs.append(fit.coef_)

plots = []
for i in range(1, p):
    plots.append(alphas)
    plots.append([c[i] for c in coeffs])

plt.plot(*plots)
plt.show()
```

```python
d = 8
a = np.linspace(0, 1, d)
b = np.hstack([[0], np.random.uniform(size=d - 2), [0]])
A = np.array([a**i for i in range(d)]).transpose()
c = np.linalg.solve(A, b)
x = np.arange(0, 1, 0.01)
X = np.array([x**i for i in range(1,d)]).transpose()
y = X @ c[1:]
plt.plot(x, y)
plt.show()
```

```python
np.random.seed(4)
n = 70
d = 5
# e = 12
train_proportion = 0.5

pol = np.poly1d(np.append(np.random.uniform(0, 1, size=d - 2), [0, 1]), r=True)
x = np.linspace(1.0/n, 1.0, n)
y = pol(x) + np.random.normal(0, 0.01, size=n)

UU = []
VV = []
for e in range(1, 10):
    U = []
    V = []
    for i in range(100):
        I = np.sort(np.random.choice(range(n), size=round(train_proportion*n), replace=False))
        J = [i for i in range(n) if i not in I]
        X = np.array([x**i for i in range(1, e + 1)]).transpose()
        b = np.linalg.pinv(X[I]) @ y[I]
#         b, rss_train, *_ = np.linalg.lstsq(X[I], y[I], rcond=-1)
        rss_train = (y[I] - X[I] @ b).dot(y[I] - X[I] @ b) 
        rss_test = (y[J] - X[J] @ b).dot(y[J] - X[J] @ b)
        U.append(rss_train)
        V.append(rss_test)
#     print(np.mean(U), np.mean(V))
    UU.append(np.mean(U))
    VV.append(np.mean(V))
# plt.plot(x, y, 'ro')
# plt.show()
```

```python
plt.plot(range(1, 10), UU)
plt.plot(range(1, 10), VV)
plt.show()
```

```python
np.random.seed(4)
n = 70
d = 5
e = 12
train_proportion = 0.5

pol = np.poly1d(np.append(np.random.uniform(0, 1, size=d - 2), [0, 1]), r=True)
x = np.linspace(1.0/n, 1.0, n)
y = pol(x) + np.random.normal(0, 0.02, size=n)

rs = np.linspace(0.00001, 0.005, 100)
UU = []
VV = []
for r in rs:
    U = []
    V = []
    for i in range(5000):
        I = np.sort(np.random.choice(range(n), size=round(train_proportion*n), replace=False))
        J = [i for i in range(n) if i not in I]
        Xt = np.array([x**i for i in range(1, e + 1)])
        X = Xt.transpose()
        M = X[I].transpose() @ X[I]
        b = np.linalg.inv(X[I].transpose() @ X[I] + r*np.identity(e)) @ X[I].transpose() @ y[I]
        rss_train = (y[I] - X[I] @ b).dot(y[I] - X[I] @ b) 
        rss_test = (y[J] - X[J] @ b).dot(y[J] - X[J] @ b)
        U.append(rss_train)
        V.append(rss_test)
    UU.append(np.mean(U))
    VV.append(np.mean(V))

```

```python
plt.plot(rs[1:250], VV[1:250])
plt.show()
```

```python
np.random.seed(4)
n = 70
d = 5
e = 20
train_proportion = 0.5

pol = np.poly1d(np.append(np.random.uniform(0, 1, size=d - 2), [0, 1]), r=True)
x = np.linspace(1.0/n, 1.0, n)
X = np.array([x**i for i in range(1, e + 1)]).transpose()
y = pol(x) + np.random.normal(0, 0.02, size=n)

alphas = np.arange(0.0001, 0.005, 0.0001)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(X, y)
print(clf.alpha_)
```

```python
plt.plot(alphas, 10**6*np.mean(clf.cv_values_, axis=0))
plt.ylabel("estimated prediction error")
plt.xlabel("λ")
plt.yticks([])
plt.title("Optimal λ from cross validation")
plt.show()
```

```python
X, y = load_diabetes(return_X_y=True)
alphas = np.arange(0.01, 2, 0.01)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(X, y)
cvs = np.mean(clf.cv_values_, axis=0)
plt.plot(alphas, cvs)
plt.ylabel("estimated prediction error")
plt.xlabel("λ")
plt.yticks([])
plt.title("Optimal λ from cross validation")
plt.show()
```

```python
X, y = load_diabetes(return_X_y=True)
XX = X - np.mean(X, axis=0) # centered
XXX = XX/np.std(XX, axis=0) # scaled
print(np.std(XXX, axis=0))
alphas = np.arange(0.1, 1000, 1)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(XXX, y)
cvs = np.mean(clf.cv_values_, axis=0)
plt.plot(alphas, cvs)
plt.ylabel("estimated prediction error")
plt.xlabel("λ")
plt.yticks([])
plt.title("Optimal λ from cross validation")
plt.show()
```
