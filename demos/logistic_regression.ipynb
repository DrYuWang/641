```python
import numpy as np
import matplotlib.pyplot as plt
from torch.autograd import Variable
import torch
```

```python
seed_ = 13
np.random.seed(seed_)
x = np.sort(np.random.uniform(size=10)*2)
y = 1.0*np.array([0, 1, 0, 0, 0, 1, 0, 1, 1, 1])
```

```python
torch.manual_seed(seed_)
X = torch.tensor(x)
Y_hat = torch.tensor(y)
A = Variable(torch.rand(1, dtype=torch.double), requires_grad=True)
B = Variable(torch.rand(1, dtype=torch.double), requires_grad=True)

# learning rate
r = 0.03

approxs = []

# descent
for _ in range(1500):
    NLL = -torch.sum(Y_hat*torch.log(torch.sigmoid(A + B*X)) + (1 - Y_hat)*torch.log(1 - torch.sigmoid(A + B*X)))
    approxs.append([A.item(), B.item(), NLL.item()])
    NLL.backward()
    
    # refine approximation
    A.data -= r*A.grad.data
    B.data -= r*B.grad.data
    
    # reinitialize gradient
    A.grad.data.zero_()
    B.grad.data.zero_()
```

```python
aa = A.item(); bb = B.item()
xs = np.arange(0, 2.5, 0.01)
ys = np.reciprocal(1 + np.exp(-(aa + bb*xs)))
plt.plot(xs, ys, 'b')
plt.plot(x, y, 'ro')
```

```python
def sigmoid(t):
    return np.reciprocal(1 + np.exp(-t))

def nll(a, b, x, y):
    return -np.array([yi*np.log(sigmoid(a + b*xi)) + (1 - yi)*np.log(1 - sigmoid(a + b*xi)) for (xi, yi) in zip(x, y)]).sum(axis=0)
```

```python
approxs = np.array(approxs)
as_ = np.arange(-5, 2.5, 0.01)
bs = np.arange(-1.5, 4, 0.01)
As, Bs = np.meshgrid(as_, bs)
Z = nll(As, Bs, x, y)
fig, ax = plt.subplots()
ax.set_aspect('equal')
CS = ax.contour(As, Bs, Z, np.flip(approxs[:,2], axis=0))
plt.plot(approxs[:,0], approxs[:,1], 'r', linestyle="solid", linewidth=3, mfc='k', mec='k')
plt.show()
```
