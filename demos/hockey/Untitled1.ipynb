```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.stats import ortho_group
import string
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
np.random.seed(13)
o = ortho_group.rvs(dim=2)
cov = o @ np.diag([2, 0.2]) @ o.T
x = np.random.multivariate_normal((0, 0), cov, size=30)
e, v = np.linalg.eigh(x.T @ x)
v = -v @ np.array([[0, 1], [1, 0]])
xv = x @ v

f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x[:,0], x[:,1], 'ro')
ax1.axis('equal')
ax1.quiver([0, 0], [0, 0], v[0,:], v[1,:], angles='xy', scale_units='xy', scale=1, color="blue")
ax1.quiver([0, 0], [0, 0], [1, 0], [0, 1], angles='xy', scale_units='xy', scale=1, color="black")
ax1.annotate("$v_1$", xy=(v[0,0], v[1,0]), xytext=(v[0,0], v[1,0] - 0.2), fontsize=24)
ax1.annotate("$v_2$", xy=(v[0,1], v[1,1]), xytext=(v[0,1] + 0.1, v[1,1]), fontsize=24)
ax1.annotate("$x_1$", xy=(1, 0), xytext=(1, 0), fontsize=24)
ax1.annotate("$x_2$", xy=(0, 1), xytext=(0, 1), fontsize=24)

# for i in range(10):
#     ax1.annotate("$" + string.ascii_uppercase[i] + "$", xy=(x[i,0], x[i,1]), xytext=(x[i,0], x[i,1] + 0.1), color="blue", fontsize=18)
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 2)

vt = v.T
ax2.plot(xv[:,0], xv[:,1], 'ro')
ax2.quiver([0, 0], [0, 0], vt[0,:], vt[1,:], angles='xy', scale_units='xy', scale=1, color="black")
ax2.quiver([0, 0], [0, 0], [1, 0], [0, 1], angles='xy', scale_units='xy', scale=1, color="blue")
# for i in [i for i in range(10)]:
#     ax2.annotate("$" + string.ascii_lowercase[i] + "$", xy=(xv[i,0], xv[i,1]), xytext=(xv[i,0], xv[i,1] + 0.1), color="blue", fontsize=18)
ax2.axis('equal')
ax2.set_xlim(-2, 2)
ax2.set_ylim(-2, 2)
plt.show()
```

```python
np.random.seed(10)
o = ortho_group.rvs(dim=3)
cov = o @ np.diag([2, 1, 0.1]) @ o.T
x = np.random.multivariate_normal((0, 0, 0), cov, size=3)
e, v = np.linalg.eigh(x.T @ x)
v = v[:, [2, 1, 0]]
xv = x @ v

f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x[:,0], x[:,1], 'ro')
ax1.axis('equal')
ax1.quiver([0, 0, 0], [0, 0, 0], v[0,:], v[1,:], angles='xy', scale_units='xy', scale=1, color="blue")
ax1.quiver([0, 0], [0, 0], [1, 0], [0, 1], angles='xy', scale_units='xy', scale=1, color="black")
ax1.annotate("$v_1$", xy=(v[0,0], v[1,0]), xytext=(v[0,0], v[1,0] - 0.2), fontsize=24)
ax1.annotate("$v_2$", xy=(v[0,1], v[1,1]), xytext=(v[0,1], v[1,1]), fontsize=24)
ax1.annotate("$v_3$", xy=(v[0,2], v[1,2]), xytext=(v[0,2] + 0.1, v[1,2]), fontsize=24)
ax1.annotate("$x_1$", xy=(1, 0), xytext=(1, 0), fontsize=24)
ax1.annotate("$x_2$", xy=(0, 1), xytext=(0, 1), fontsize=24)
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 3)

vt = v.T
ax2.plot(xv[:,0], xv[:,1], 'ro')
ax2.quiver([0, 0, 0], [0, 0, 0], vt[0,:], vt[1,:], angles='xy', scale_units='xy', scale=1, color="black")
ax2.quiver([0, 0], [0, 0], [1, 0], [0, 1], angles='xy', scale_units='xy', scale=1, color="blue")
ax2.axis('equal')
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
plt.show()
```
