## Simple linear regression via gradient descent

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (10.0, 10.0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from torch.autograd import Variable
import torch
```

```python
# Generate data.
seed_ = 666
np.random.seed(seed_)
n = 10
s = 0.25
a = np.random.uniform()
b = np.random.uniform()
x = np.random.uniform(size=n)
y = a + b*x + s*np.random.normal(size=x.size)

# Compute the regression coefficients using standard formulas.
xbar = np.mean(x)
ybar = np.mean(y)
Sxx = sum((x - xbar)*(x-xbar))
Sxy = sum((x-xbar)*(y-ybar))
ahat = ybar - Sxy/Sxx*xbar
bhat = Sxy/Sxx
```

```python
# initialize
torch.manual_seed(seed_)
A = Variable(torch.rand(1, dtype=torch.double), requires_grad=True)
B = Variable(torch.rand(1, dtype=torch.double), requires_grad=True)
xx = torch.tensor(x)
yy = torch.tensor(y)
loss = torch.nn.MSELoss()

# learning rate
r = 0.03
approxs = []

# descent
for _ in range(1000):
    Y = A + B*xx
    MSE = loss(Y, yy)
    approxs.append([A.item(), B.item(), MSE.item()])
    MSE.backward()
    
    # refine approximation
    A.data -= r*A.grad.data
    B.data -= r*B.grad.data
    
    # reinitialize gradient
    A.grad.data.zero_()
    B.grad.data.zero_()
    
    # store refined approximation

approxs = np.array(approxs)
aa = A.item()
bb = B.item()
```

```python
plt.plot(np.array([0, 1]), aa + bb*np.array([0, 1]), 'k')
plt.plot(np.array([0, 1]), ahat + bhat*np.array([0, 1]), 'm')
plt.plot(x, y, 'bo')
plt.show()
```

```python
def mse(As, Bs, x, y):
    return np.array([np.square(As + xi*Bs - yi) for (xi, yi) in zip(x, y)]).sum(axis=0)/n
```
