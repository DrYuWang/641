```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
def G(x, mu=0, var=1):
    return 1./np.sqrt(2.*np.pi*var)*np.exp(-np.power(x - mu, 2.) / (2. * var))

def em(x, r_init, T):
    K = len(r_init)
    r_ = r_init.copy()
    rs_ = [r_]
    pis_ = []
    mus_ = []
    vars_ = []
    for t in range(T):
        R = r_.sum(axis=1)
        pi_ = R/N
        mu_ = np.reshape(r_.dot(x)/R, (K, 1))
        var_ = np.sum(r_ * np.square(np.array([x for k in range(K)]) - mu_), axis=1)/R
        r_ = np.array([pi_[k]*G(x, mu_[k], var_[k]) for k in range(K)])
        r_ = r_/r_.sum(axis=0)
        pis_.append(pi_)
        mus_.append(mu_[:,0])
        vars_.append(var_)
        rs_.append(r_)
    return {"pi_": pi_, "mu_": mu_, "var_": var_, "r_": r_, "pis_": np.array(pis_),
            "mus_": np.array(mus_), "vars_": np.array(vars_), "rs_": rs_}
```

```python
np.random.seed(680)

K = 2
mu = np.array([1, 4])
var = np.array([0.5, 0.75])
n = np.array([50, 150])
N = n.sum() # compensate for possible rounding error
pi = n/N
x = np.sort(np.concatenate([np.random.normal(mu[k], np.sqrt(var[k]), size=n[k]) for k in range(K)]))

M = np.floor(N/K).astype(int)
r_ = np.zeros((K, N))
r_[0, :M] = 1
r_[1, M:] = 1

# r_ = np.random.uniform(size=(K, N))
# r_ = r_/r_.sum(axis=0)

fit = em(x, r_, 50)
pi_ = fit["pi_"]; mu_ = fit["mu_"]; var_ = fit["var_"]; r_ = fit["var_"]

X = np.arange(-1, 7, 0.1)
Y = sum([pi[k]*G(X, mu[k], var[k]) for k in range(K)])
Y_ = sum([pi_[k]*G(X, mu_[k], var_[k]) for k in range(K)])
plt.figure(figsize=(8, 6))
plt.plot(X, Y, 'r')
plt.plot(X, Y_, 'c')
plt.hist(x, bins=35, density=True)
plt.plot()
plt.show()
```
