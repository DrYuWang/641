```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines
from scipy.interpolate import lagrange
```

```python
xs = np.arange(-1, 1, 0.1)
ys = [1/(1+25*x*x) for x in xs]
def y(x): return 1/(1+25*x*x)
```

```python
plt.plot(xs, ys)
```

```python
n = 10
xsn = np.arange(-1, 1.00001, 2/n)
pn = lagrange(xsn, [y(x) for x in xsn])
plt.plot(*[xs, [pn(x) for x in xs]])
```
