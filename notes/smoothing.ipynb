```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)
y0 = 1/(1+15*x*x) + 0.05*np.random.randn(100)
y1 = np.append([0], y0[:-1])
y2 = np.append([0], y1[:-1])
y3 = np.append([0], y2[:-1])
y4 = np.append([0], y3[:-1])
y = 1/5*(y0 + y1 + y2 + y3 + y4)
plt.plot(x, y, x, y0)
```
