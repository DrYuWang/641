```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines
from scipy.interpolate import interp1d, lagrange, barycentric_interpolate
from math import factorial as fact
from math import pi, sin
```

```python
x = np.linspace(0, 1, 10)
xx = np.linspace(0, 1, 100)
(m, n) = (2, 3)
f, ax = plt.subplots(m, n, sharex=True, sharey=True, figsize=(20, 9/16*20))
for i in range(m):
    for j in range(n):
        y = x**4 + 0.075*np.random.randn(len(x))
        yy = barycentric_interpolate(x, y, xx)
        ax[i, j].plot(xx, yy, xx, xx**4, x, y, 'o')
plt.savefig("grid.svg")
```
