```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import scipy.stats as stats
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import scale
```

```python
np.arange(10)**2 + np.random.normal(0, 1, 10)
```

```python
x = np.arange(0, 10, 0.2)
y = x**6/100000 + np.random.normal(0, 1, len(x))
m, b, *_ = stats.linregress(x, y)
yhat = m*x + b
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.plot(x, y, 'ro', x, x**6/100000, 'b', x, yhat, 'k')
plt.show()
```

```python
x1, x2, y1, y2 = tts(x, y, test_size=0.5)
```

```python
m1, b1, *_ = stats.linregress(x1, y1)
y1hat = m1*x1 + b1
e1 = y1 - y1hat
RSS1 = np.dot(e1, e1)
TSS1 = np.dot(y1 - np.mean(y1), y1 - np.mean(y1))
```

```python
y2hat = m1*x2 + b1
e2 = y2 - y2hat
RSS2 = np.dot(e2, e2)
TSS2 = np.dot(y2 - np.mean(y2), y2 - np.mean(y2))
```

```python
print(RSS1, RSS2)
```
